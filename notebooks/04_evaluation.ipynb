```python
import pandas as pd
import joblib
```

```python
# Load model
model = joblib.load("C:\\Users\\Convenience\\Downloads\\CRYPTO\\crypto-classifier\\models\\buy_sell_classifier.pkl")
```

```python
# Load features
df = pd.read_csv("C:\\Users\\Convenience\\Downloads\\CRYPTO\\crypto-classifier\\data\\processed\\BTCUSDT_features.csv")
X = df.drop(columns=["open_time"], errors="ignore")
```

```python
# Predict
# Select only the features the model was trained on
X_pred = X[['open', 'high', 'low', 'close', 'volume', 'return_1d', 'return_7d', 'volatility_7d', 'rsi', 'sma_20', 'sma_50', 'sma_200', 'macd']]
df["Prediction"] = model.predict(X_pred)
```
